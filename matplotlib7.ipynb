# Box Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,notch=True) #to change the shape
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,vert=False) #to make shape horizontal
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,widths=0.4,labels=['python'],patch_artist=True) #to change the width #labels=['python'] to give label to x-axis  #patch_artist=True is use for to fill color in box plot
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=['python'],showmeans=True) #showmeans=True to see middle,mean or Average of box plot
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True) #showmeans=True to see middle,mean or Average of box plot
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,whis=2) #to join outler to line whis=2
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,sym='g+') #to change color and sign of outler use  sym='g+'
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,sym=' ') #to hide outler make sym=' '
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='b')) #to boxprops=dict(color='r') to change the color of box #to capprops=dict(color='b') to change color of line diff side 
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),whiskerprops=dict(color='y'),capprops=dict(color='b'))#to change color of whisk whiskerprops=dict(color='y') #to boxprops=dict(color='r') to change the color of box #to capprops=dict(color='b') to change color of line diff side 
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),whiskerprops=dict(color='y'),capprops=dict(color='b'),flierprops=dict(markeredgecolor='y'))#to change color of whisk whiskerprops=dict(color='y') #to boxprops=dict(color='r') to change the color of box #to capprops=dict(color='b') to change color of line diff side 
plt.show()
```

```python
#to make multiple boxplot
x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,90]

y=[x,x1]
plt.boxplot(y,labels=['python','C++'],showmeans=True,sym='g+',boxprops=dict(color='r'),whiskerprops=dict(color='y'),capprops=dict(color='b'),flierprops=dict(markeredgecolor='y'))
plt.show()
```
