# text in matplotlib

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.xlabel("language",fontsize=15)# to give font size
plt.ylabel("No",fontsize=15)# to give font size
plt.title('Vishwa',fontsize=15) 


plt.text(2,5,"java",fontsize=15,style='italic',bbox={'facecolor':'yellow'}) 
#to point something at(2,5)point 
#style='italic' to make text italic 
#to give color use bbox={'facecolor':'yellow'}



plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.xlabel("language",fontsize=15)# to give font size
plt.ylabel("No",fontsize=15)# to give font size
plt.title('Vishwa',fontsize=15) 


plt.text(2,5,"java",fontsize=15,style='italic',bbox={'facecolor':'yellow'}) 
#to point something at(2,5)point 
#style='italic' to make text italic 
#to give color use bbox={'facecolor':'yellow'}


plt.annotate('python',xy=(1,2),xytext=(4,4),arrowprops=dict(facecolor='pink',shrink=50))  #to annotate something or some point with arrow

plt.legend(['up'],loc=2,facecolor='blue',edgecolor='green',framealpha=0.5,shadow=True) 
#to give egde golor around box 
#framealpha=0.5 to make it transparent
#shadow to give shadow around box

plt.show()
```
