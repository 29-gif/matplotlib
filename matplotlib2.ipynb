# scatterplot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

plt.xlabel("Day",fontsize=15)# to give x label and font size
plt.ylabel("No",fontsize=15)# to give y label and font size
plt.title('Scatter Plot',fontsize=15) # to give title to chart and font size

plt.scatter(day,no,color='y')#to give color to bar chart
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['y','b','m','g','r','g','r']

plt.xlabel("Day",fontsize=15)# to give x label and font size
plt.ylabel("No",fontsize=15)# to give y label and font size
plt.title('Scatter Plot',fontsize=15) # to give title to chart and font size

plt.scatter(day,no,color=c)#to give color to bar chart
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['y','b','m','g','r','g','r'] #list for colors
sizes=[100,120,200,300,60,180,500] #list for sizes

plt.xlabel("Day",fontsize=15)# to give x label and font size
plt.ylabel("No",fontsize=15)# to give y label and font size
plt.title('Scatter Plot',fontsize=15) # to give title to chart and font size

plt.scatter(day,no,color=c,s=sizes)#to give size to scatter
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['y','b','m','g','r','g','r'] #list for colors
sizes=[100,120,200,300,60,180,500] #list for sizes

plt.xlabel("Day",fontsize=15)# to give x label and font size
plt.ylabel("No",fontsize=15)# to give y label and font size
plt.title('Scatter Plot',fontsize=15) # to give title to chart and font size

plt.scatter(day,no,color=c,s=sizes,alpha=0.3) #to blur a graph use alpha (alpha value should be btw 0 to 1)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['y','b','m','g','r','g','r'] #list for colors
sizes=[100,120,200,300,60,180,500] #list for sizes

plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Scatter Plot',fontsize=15) 

plt.scatter(day,no,color=c,s=sizes,alpha=0.1,marker="*") #give marker parameter to change shapes of scatter
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['y','b','m','g','r','g','r'] #list for colors
sizes=[600,900,200,300,60,180,500] #list for sizes

plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Scatter Plot',fontsize=15) 

plt.scatter(day,no,color=c,s=sizes,marker="*",edgecolor='g',linewidth=1) #to change edge color #to give width to line
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,40,70,56,20,30] #list for colors # #you can use code aswell,but it should be between 0 to 100 only
sizes=[600,900,200,300,60,180,500] #list for sizes

plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Scatter Plot',fontsize=15) 

plt.scatter(day,no,s=sizes,cmap="viridis",c=colors) #to change edge color #to give width to line
#cmap is color from where it going to pick,there are so many different names aswell than viridis for example(accent,blues,brbg,bugn,bupu,dark2,gnbu,orrd) #CHECK ON GOOGLE
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,40,70,60,20,30]
sizes=[600,900,200,300,60,180,500] 

plt.scatter(day,no,s=sizes,cmap="BrBG",c=colors) 

plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Scatter Plot',fontsize=15)

T=plt.colorbar() #to see color bar
T.set_label("ColorBar",fontsize=15) #to give colorbar label


plt.show()
```

```python
#to show two diffrnt graphs together ,mulpitple plots together
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,5,4,1,6,2]

colors=[10,49,40,70,60,20,30]
sizes=[600,900,200,300,60,180,500] 

plt.scatter(day,no,s=sizes,cmap="BrBG",c=colors) 
plt.scatter(day,no2,s=sizes,color='b') 

plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Scatter Plot',fontsize=15)

T=plt.colorbar() #to see color bar
T.set_label("ColorBar",fontsize=15) #to give colorbar label


plt.show()
```
