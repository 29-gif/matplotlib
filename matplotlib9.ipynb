# stack plot or area plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]

plt.stackplot(x,area)

plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,3,2,5,4]

plt.stackplot(x,area)

plt.show()
```

```python
#to make multiple stackplot
x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

plt.stackplot(x,area,area1,area2)

plt.show()
```

```python

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=['area','area1','area2']

plt.stackplot(x,area,area1,area2,labels=l)#to show labels 

plt.legend(loc=2)#to show labels #to change location of label
plt.show()
```

```python

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=['area','area1','area2']

plt.stackplot(x,area,area1,area2,labels=l,colors=['r','g','y'])#to change color

plt.legend(loc=2)
plt.show()
```

```python

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=['area','area1','area2']

plt.stackplot(x,area,area1,area2,labels=l,colors=['r','g','y'],baseline='sym')
#to change baseline
#if baseline= zero(will not make any diff) 
#if baseline= sym (will remove baseline) 
#if baseline= wingle

plt.title('python')# to give title to chart
plt.xlabel("x-axis")# to give x label
plt.ylabel("y-axis")# to give y label

plt.grid()#to check grid
plt.legend(loc=2)
plt.show()
```
