# Sub Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,color='r') #linear plot
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.pie([1],colors='r') 


plt.subplot(2,2,2)
plt.plot(x,y,color='r') #linear plot


x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x1)


x2=['a','s','d','f']
y2=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show()
```
