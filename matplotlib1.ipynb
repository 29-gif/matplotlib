# Bar Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language")# to give x label
plt.ylabel("No")# to give y label
plt.title('Vishwa') # to give title to chart

plt.bar(x,y)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)# to give font size
plt.ylabel("No",fontsize=15)# to give font size
plt.title('Vishwa',fontsize=15) 

plt.bar(x,y)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

plt.bar(x,y,width=0.4) #to give width to bar chart
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

plt.bar(x,y,width=0.4,color='y') #to give color to bar chart
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

c=['y','b','m','g'] #to give diffrent colors to every bar 

plt.bar(x,y,width=0.4,color=c)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

c=['y','b','m','g'] 

plt.bar(x,y,width=0.4,color=c,align='edge') #to put all languages or x axis values to certain position from centre #align has two parameters(edge|center)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

c=['y','b','m','g'] 

plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=4,linestyle=':') #to give color to border use edgecolor 
#to give width to border use linewidth #to give styling to line use linestyle

plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

c=['y','b','m','g'] 

plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=4,linestyle=':',alpha=0.3) #to blur a graph use alpha (alpha value should be btw 0 to 1)
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 

c=['y','b','m','g'] 

plt.bar(x,y,width=0.4,color=c,edgecolor='r',label='highest value of language') #to give label to bar chart #when u use always use plt.legend()
plt.legend()
plt.show()
```

```python
#to make multiple bar graphs
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,70,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 


plt.bar(x,y,width=0.4,color='r',edgecolor='r',label='highest value of language')
plt.bar(x,z,width=0.4,color='y',edgecolor='y',label='highest value of language 2')
plt.legend()
plt.show()
```

```python
#to show two diff barcharts
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,70,50]

width=0.1
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 


plt.bar(p,y,color='r',edgecolor='r',label='highest value of language')
plt.bar(p1,z,color='y',edgecolor='y',label='highest value of language 2')


plt.legend()
plt.show()
```

```python
#to show two diff barcharts
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,70,50]

width=0.1
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 


plt.bar(p,y,color='r',edgecolor='r',label='highest value of language')
plt.bar(p1,z,color='y',edgecolor='y',label='highest value of language 2')

plt.xticks(p+width/2,x) #to show names of x axis
plt.legend()
plt.show()
```

```python
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,70,50]

width=0.1
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 


plt.bar(p,y,color='r',edgecolor='r',label='highest value of language')
plt.bar(p1,z,color='y',edgecolor='y',label='highest value of language 2')

plt.xticks(p+width/2,x,rotation=20) #to show names of x axis to tilt of any angle
plt.legend()
plt.show()
```

```python
# to show horizontal bar graph
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,70,50]



plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.title('Vishwa',fontsize=15) 


plt.barh(x,y,color='r',label='highest value of language')
plt.barh(x,z,color='y',label='highest value of language 2')


plt.legend()
plt.show()
```
