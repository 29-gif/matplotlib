# step plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5]
y=[1,2,5,4,5]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,2,5,3,4]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,2,5,3,4]

plt.step(x,y,marker='o',ms=10,mfc='g',color='r')#to give marker from where its going to change marker='o' #ms=10 to give size to marker #to change marker color mfc

plt.xlabel("x-axis")# to give x label
plt.ylabel("y-axis")# to give y label
plt.title('Python') # to give title to chart
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,2,5,3,4]

plt.step(x,y,marker='o',ms=10,mfc='g',color='r',label="python")#to give python label to chart

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Python') 

plt.legend(loc=2)#to change location of chart #it could be 1,2,3,4
plt.grid()
plt.show()
```
