# Pie Chart

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']

plt.pie(x) 
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']

plt.pie(x,labels=y) #to see labels which one is which
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=px)  #to see some part outof the pie chart
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] #to give color parameter

plt.pie(x,labels=y,explode=px,colors=c)  
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%")  #to see percentage of each pie portion use autopct="%0.1f%%"
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.2f%%")  #if autopct="%0.2f%%" it will saw two 00 after decimal
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.2f%%",shadow=True,radius=1.5)  #to show shadow in it #to make it big give radius
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.2f%%",shadow=True,radius=1.5,labeldistance=1.5)  #to give label a distance 
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.1,startangle=180)  #to give starting angle (bydefault=0, you can use ,90,45,180,360)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12})  #to give font size to all the x and y text #it shoulbe in dictionary 
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=False)  #to change the the direction of diagram check above and below for diff(c,c++)
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,
        radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=False,wedgeprops={'linewidth':5}) #to Increase line width or shadow #it shoulbe in dictionary  
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,
        radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=False,wedgeprops={'linewidth':5,"edgecolor":'r'}) #to give color to shadow
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,
        radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=False,wedgeprops={'linewidth':5}
       ,center=(1,3)) #to give center 
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,
        radius=1.5,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=False,wedgeprops={'linewidth':5}
       ,center=(1,3),rotatelabels=True) #to rotate labels 
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
px=[0.3,0.0,0.0,0.0]
c=['y','b','m','g'] 

plt.pie(x,labels=y,explode=px,colors=c,autopct="%0.1f%%",shadow=True,
        radius=1.1,labeldistance=1.1,startangle=180,
       textprops={"fontsize":12},counterclock=True,wedgeprops={'linewidth':5}
       ) 
plt.title("Languages")#to give title to graph
plt.legend(loc=2)#to change location of legend
plt.show()
```

```python
#to make dotpie chart
plt.pie([1])
plt.show()
```

```python
#to make donatpie chart

x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','Python']
c=['r','b','g','y']

plt.pie(x,labels=y,radius=1.5 ) 
plt.pie(x1,radius=0.5,colors=c ) #to make multiple graphs

plt.title("Languages")
plt.legend(loc=2)
plt.show()
```

```python
#to make donatpie chart

x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','Python']


plt.pie(x,labels=y,radius=1.5 ) 
plt.pie([1],colors='b' ) #can make ring as well

plt.title("Languages")
plt.legend(loc=2)
plt.show()
```

```python


x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','Python']


plt.pie(x,labels=y,radius=1.5 ) 
#orrrrrrr
cr=plt.Circle(xy=(0,0),radius=1,facecolor='b')
plt.gca().add_artist(cr)

plt.title("Languages")
plt.legend(loc=2)
plt.show()
```
