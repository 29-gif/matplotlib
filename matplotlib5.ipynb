# Stem plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.') #to change line #itcould be one of this(:   -.  --  -)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="r+") #to change circle and color (b+  g+ r+ ro)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="r+",bottom=3) #to start line bottom line from specific number
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="r+",bottom=9) #to start line bottom line from specific number
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="r+",basefmt='g') #to change base line color
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="ro",basefmt='g',label='python') #to saw label of line

plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt='-.',markerfmt="ro",basefmt='g',label='python',
         use_line_collection=False,orientation='horizontal') #to change line colors #to show graph horizontal

plt.legend()
plt.show()
```
